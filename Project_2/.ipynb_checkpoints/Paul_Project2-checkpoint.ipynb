```python
##################################################################################
################## clean column names ##########################################
#################################################################################


columns1 = ['year', 'artist', 'track', 'time', 'genre', 'date_entered', 'date_peaked']

weekname = ['week_' + str(i) for i in range(1,77)]
weekname
columns = columns1+weekname

columns

df.columns = columns


##################################################################################
################## clean column names ##########################################
#################################################################################


df.head()

```

```python
############################################################
##################### run once ##########################
########################################################

del df['year']
df.head()

############################################################
##################### run once ##########################
########################################################
```

```python
##################################################################################
################## clean week1 from an int to a float #########################
#################################################################################
df['date_entered'] = pd.to_datetime(df['date_entered'])

df['date_peaked'] = pd.to_datetime(df['date_peaked'])

df.head()
##################################################################################
################## turn date into datetime #########################
#################################################################################
   
```

```python
##################################################################################
################## clean week1 from an int to a float #########################
#################################################################################

df['week_1'] = df['week_1'].astype(float)
df.head()


##################################################################################
################## clean week1 from an int to a float #########################
#################################################################################
```

```python
##################################################################################
##################       create field time to peak    #########################
#################################################################################


df['time_to_peak'] = df['date_peaked'] - df['date_entered']

df.head()


```

```python
###############################################################################################
################################ GENERATES LIST OF DATES ######################################
###############################################################################################

import datetime 

MainDate= []

start = datetime.datetime.strptime('7-1-1999', '%m-%d-%Y')
end = datetime.datetime.strptime('1-1-2001', '%m-%d-%Y')

date_generated = [start + datetime.timedelta(days=x) for x in range (0, (end-start).days)]

for date in date_generated:
    MainDate.append(date.strftime('%m-%d-%Y'))

###############################################################################################
################################ GENERATES LIST OF DATES ######################################
###############################################################################################
```

```python
#################################################################################################
################## create pandas dataframe dateDF that will be the index ##########################
#######################################################################################################

dateDF = pd.DataFrame(MainDate, columns = ['date_entered'])
dateDF.head()


#################################################################################################
################## create pandas dataframe dateDF that will be the index ##########################
#######################################################################################################

```

```python
#################################################################################################
###################################### convert to datetime        ######################################
#######################################################################################################

dateDF['date_entered'] = pd.to_datetime(dateDF['date_entered'])

dateDF.head()

#################################################################################################
###################################### convert to datetime        ######################################
#######################################################################################################
```
