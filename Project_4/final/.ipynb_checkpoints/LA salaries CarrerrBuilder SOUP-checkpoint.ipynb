```python
import requests
from bs4 import BeautifulSoup

URL = 'http://www.careerbuilder.com/jobs-data-in-los-angeles?&pay=20'
#END_URL = 

## get page1
r = requests.get(URL)
soup = BeautifulSoup(r.text, 'lxml')
```

```python
soup.find('div', attrs={"class": "column"})
```

```python
myDict = []
for item in soup.select('div.job-row'):
    myDict.append(item.get_text(strip=True))
```
