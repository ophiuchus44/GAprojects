```python
def generatePositionLevel(title):
    title = title.lower()
    if (title.find('manager',0,len(title)) >= 0) or (title.find('management',0,len(title)) >= 0) or (title.find('python',0,len(title)) >= 0) or (title.find('data scientist',0,len(title)) >= 0) or (title.find('analyst',0,len(title)) >= 0):
        return 'middle'
    elif (title.find('senior',0,len(title)) >= 0) or (title.find('machine learning',0,len(title)) >= 0) or (title.find('principal',0,len(title)) >= 0) or (title.find('chief',0,len(title)) >= 0) or (title.find('vp',0,len(title)) >= 0) or (title.find('director',0,len(title)) >= 0) or (title.find('sr',0,len(title)) >= 0):
        return 'upper'
    return 'other'
    
df['position_level'] = df['title'].apply(generatePositionLevel)
df
```

```python
Xs = patsy.dmatrix('city + position_level -1', df)
Ys = Y
```

```python
logit2 = lm.fit(Xs, Ys)
```

```python
Xs_train, Xs_test, Ys_train, Ys_test = train_test_split(Xs, Ys, test_size=0.33, stratify = Ys, random_state = 15)
```

```python
logit2.fit(Xs_train, Ys_train)
```

```python
predictions2 = logit2.predict(Xs_test)
print predictions2
```
